### **Part 4: Classification**
Training and Testing Decision Trees

**Create class label**

```python
# Create class label for your assigned variable (based on mean value)
def create_class_label (row):
    mean_value = dataset['Sale Amount'].mean()
    if row['Sale Amount'] > mean_value:
        return 1 # value of assigned variable > mean value (high)
    return 0 # value of assigned variable <= mean value (low)
```

```python
mean_value = dataset['Sale Amount'].mean()
dataset['Class Label'] = (dataset['Sale Amount'] > mean_value).astype(int)
dataset.head()
```

**Divide the dataset into a training set (75% of the data) and a test set (25% of the data).**

```python
# Partition dataset into training and test sets
predictors_train, predictors_test, class_train, class_test =  train_test_split(dataset[['List Year', 'Assessed Value', 'Sale Amount']],  # predictors
                                                              dataset['Class Label'],           # class
                                                              test_size = 0.25,                 # test size
                                                              random_state = 0)
# Try different combinations of predictors!
```

**<u>Using the training set</u>, build a decision tree to predict the class label based on the predictors.**

```python
# Build decision tree
model = tree.DecisionTreeClassifier(random_state = 0)
model.fit(predictors_train, class_train) # Train decision tree
```

```python
# Plot decision tree
plt.figure(figsize = (15, 15)) # Set figure size (if needed)
tree.plot_tree(model, feature_names = ['Assessed Value', 'Town', 'Sale Amount'],      # feature_names indicates the names of the predictors
               class_names = ['Low', 'High'],     # class_names indicates the names of the classes
               filled = True, impurity = False)
```

**Evaluate the performance of this decision tree <u>on the test set</u>.**

```python
# Predict class labels for test set
class_pred = model.predict(predictors_test) # Use decision tree to predict class labels
print(class_pred) # Print predicted class labels
```

```python
# Plot confusion matrix
conf_matrix = metrics.confusion_matrix(class_test, class_pred) # Create confusion matrix
sns.heatmap(conf_matrix, annot = True, fmt = ".3f", square = True, cmap = plt.cm.Blues) # Plot confusion matrix
plt.ylabel('True') # Set y-axis label
plt.xlabel('Predicted') # Set x-axis label
plt.title('Confusion matrix') # Set chart title
plt.tight_layout()
```

```python
# Compute evaluation metrics
print(metrics.accuracy_score(class_test, class_pred)) # Compute accuracy
print(1 - metrics.accuracy_score(class_test, class_pred)) # Compute error
```
